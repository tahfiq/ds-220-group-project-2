```python
import datascience as ds

import matplotlib
matplotlib.use('Agg')
%matplotlib inline
import matplotlib.pyplot as plots

import numpy as np
import pandas as pd
```

```python
EV = ds.Table.read_table('Electric_Vehicle_Population_Data.csv')
```

```python
EV
```

```python
make_arr=EV['Make']
```

```python
total_count=0
for i in make_arr:
  total_count+=1

print(total_count)
```

```python
list_of_cars=[]

for i in make_arr:
  if i not in list_of_cars:
    list_of_cars.append(i)

print(list_of_cars)
len(list_of_cars)
```

```python
def countmake(x):
  count=0
  for i in make_arr:
    if i==x:
      count+=1
  return count
```

```python
count_dict={}

for i in list_of_cars:
      a=countmake(i)
      count_dict[i]=a

sorted(count_dict.items())
```

```python
sorted_count=sorted(count_dict.items())
xaxis,yaxis=zip(*sorted_count)
fig, ax = plots.subplots(1, figsize = (20, 8))
ax.grid()
fig.autofmt_xdate()
plots.plot(xaxis,yaxis)
plots.show

```

```python
most_common_make = df['Make'].value_counts().idxmax()

print("The most common car make is:", most_common_make)
```

```python
tesla_wa= df[(df['Make'] == 'TESLA')]
most_popular_tesla_model = tesla_wa['Model'].value_counts().idxmax()

print("The most popular Tesla model in Washington:", most_popular_tesla_model)
```

```python
electric_car_range_df = df[df['Electric Range'] <= 100]

print("Cars with an electric range of 0-100:")
print(electric_car_range_df)
```

```python
electric_car_range=df.loc[df['Electric Range'] <= 100]
most_common_county = electric_car_range['County'].value_counts().idxmax()

print("The cars with an electric range of 0-100 are most common in the county:", most_common_county)
```

```python
newer_cars=df.loc[df['Model Year'] >= 2020]
in_postal_code = newer_cars['Postal Code'].value_counts().idxmax()
print("The postal code with the newest cars:", in_postal_code)
```

```python
electric_car_range2 = df.loc[df['Electric Vehicle Type'] == 'Plug-in Hybrid', 'Electric Range'].mean()

print("The average electric range for plug-in hybrid electric vehicles is:", electric_car_range2)



df['Electric Range'] = pd.to_numeric(df['Electric Range'], errors='coerce')
phev_average_range = df.loc[df['Electric Vehicle Type'] == 'Plug-in Hybrid', 'Electric Range'].mean()

print("The average electric range for plug-in hybrid electric vehicles is:", phev_average_range)
```
